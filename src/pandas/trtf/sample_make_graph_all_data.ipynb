```python
import csv
from collections import OrderedDict, defaultdict

import matplotlib.pyplot as plt

import pandas as pd

yaxis = {"tr": [40.0, 70.0], "tf": [40.0, 70.0], "duty": [40.0, 60.0]}
ylabels = {"tr": "ps", "tf": "ps", "duty": "%"}

index = []
data = defaultdict(list)
with open("data.csv") as f:
    for row in csv.reader(f):
        index.append(row[0])
        for i in range(1, len(row), 2):
            data[row[i]].append(float(row[i + 1]))

for val in data:
    df = pd.DataFrame(data[val], index=index, columns=[val])
    fig = plt.figure(figsize=(8, 4))
    ax = fig.add_subplot(1, 1, 1)
    ax.yaxis.set_major_formatter(plt.FormatStrFormatter("%.1f"))
    ax.set_ylabel(ylabels[val], fontsize=20)
    
    df.plot(ax=ax, ylim=yaxis[val], style=["bo"], legend=True, fontsize=20)
    ax.legend(fontsize=20)
    ax.grid(axis="y")
    print(df)
    plt.savefig(val + ".png")
#     plt.close("all")
```
