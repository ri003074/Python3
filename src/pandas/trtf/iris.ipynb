```python
import csv

import matplotlib.pyplot as plt

import pandas as pd
```

```python
df = pd.read_csv("iris.csv")
```

```python
df
```

```python
fig = plt.figure()
df.plot()
```

```python
fig = plt.figure()  # subplotを入れるfigure図の全体を作成する
ax = fig.add_subplot(1, 1, 1)  # figにsubplotを追加　1行1列の位置1に置く
print(type(fig))  # matplotlib.figure.Figure
# df.plot()  # 図を入れる　アクティブなサブプロットに描画される
df.plot(ax=ax)  # こっちでもOK
# print(type(df.plot()))  # matplotlib.axes._subplts.AxesSubplot
fig.savefig("iris.png")
# fig.close("all")
```

```python
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(9, 6))
df.plot(ax=axes[0, 0], legend=False)
df.plot(ax=axes[1, 2], legend=False, kind="hist")
fig.savefig("iris_subplots.png")
```
