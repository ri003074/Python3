```python
import matplotlib.pyplot as plt
import numpy as np

import pandas as pd

fig, ax = plt.subplots()

# hide axes
fig.patch.set_visible(False)
ax.axis("off")
ax.axis("tight")

df = pd.DataFrame(np.random.randn(10, 4), columns=list("ABCD"))

ax.table(cellText=df.values, colLabels=df.columns, loc="center")

fig.tight_layout()

plt.show()
```
