```python
import csv
from collections import OrderedDict

import matplotlib.pyplot as plt

import pandas as pd

index = []
tr = []
tf = []
with open("data.csv", "r") as f:
    for row in csv.reader(f):
        index.append(row[0])
        tr.append(int(row[2]))
        tf.append(int(row[4]))

data = OrderedDict(tr=tr, tf=tf)

df = pd.DataFrame(data, index=index)

plt.figure()
df.plot(
    figsize=(10, 5),
    kind="line",
    legend=True,
    grid=True,
    style=["bo", "ro"],
    subplots=True,
)
plt.savefig("a.png")
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(9, 6))
```

```python
df.plot(ax=axes[0, 1], legend=True)
```
