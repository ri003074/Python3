```python
import numpy as np
from PIL import Image

import matplotlib.pyplot as plt
```

## matplotlibにはグラフを作る際の2つの流儀がある

### オブジェクト指向インターフェース 

```python
fig, ax = plt.subplots()  # figureオブジェクトとそれに属する１つのAxesオブジェクトを同時に作成
ax.plot()

# or

fig = plt.figure()  # figureオブジェクトを作成
ax = fig.add_subplot(1, 1, 1)  # figに属するaxesオブジェクトを作成
ax.plot()
```

### fig, axes, axisの関係 

```python
file_name = "fig_axes_axis.png"
array_obj = plt.imread(file_name)
plt.imshow(array_obj)
```

```python
fig = plt.figure()  # figureオブジェクトを作成
axes = fig.add_subplot(1, 1, 1)  # figに属するaxesオブジェクトを作成
axes.plot()

print("fig.axes: ", fig.axes)
```

```python
print("axes.figure: ", axes.figure)
```

```python
print("axes.xaxis: ", axes.xaxis)
```

```python
print("axes.yaxis: ", axes.yaxis)
```

```python
print("axes.xaxis.axes: ", ax.xaxis.axes)
```

```python
print("axes.yaxis.axes: ", ax.yaxis.axes)
```
